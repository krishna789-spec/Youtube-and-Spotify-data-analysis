HOW DO AVERAGE LIKES,VIEWS AND COMMENT ARE COMPARED BETWEEN DIFFRENT ALBUM TYPES

```python
data.head(1)
```

```python
avg=data.groupby('Album_type')[['Likes','Views','Comments']].mean() # gave the mean of all album type rows with like,views and comment 
avg
```

```python
type(avg)
```

```python
#to convert  index of the series  to the  colummn of the dataframe 
avg=avg.reset_index()
avg
```

```python
#melt the dataframe to show in bar plot melt - unpoivt a data frame
avg_melted = pd.melt(avg,id_vars='Album_type',var_name="Attribute",value_name="Total")
avg_melted
```

```python
#draw a plot 
plt.Figure(figsize=(9,4))
sns.barplot(x="Album_type",y="Total",hue="Attribute",data=avg_melted)
plt.title('Avg likes view and comment')
plt.xlabel('Album Type')
plt.ylabel('total')
plt.show()
```
